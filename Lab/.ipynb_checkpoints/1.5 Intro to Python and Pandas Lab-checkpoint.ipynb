```python
import pandas as pd
df = pd.read_csv('netflix_titles.csv')
df.head()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['year_added'].value_counts().sort_index()
```

```python
df['year_added'].value_counts().sort_index().plot(kind='bar', figsize=(10,5), title='Number of Titles Added Per Year')
```

**How has Netflix’s catalog grown over time?**

The graph shows that Netflix’s catalog grew steadily over the 2010s, with a sharp increase after 2015 as Netflix expanded globally and invested heavily in original content. Growth slowed in recent years, possibly reflecting a shift toward quality over quantity.

```python
df['listed_in'].str.split(', ').explode().value_counts().head(10)
```

```python
df['listed_in'].str.split(', ').explode().value_counts().head(10).plot(kind='bar', figsize=(10,5), title='Top 10 Genres on Netflix')
```

**What genres are most common on Netflix?**

The most common categories include International Movies, Dramas, and Comedies. This suggests that Netflix’s catalog leans heavily toward globally appealing and story-driven content. I love my asian dramas like Ice Fantasy.

```python
df['country'].value_counts().head(10)
```

```python
df['country'].value_counts().head(10).plot(kind='bar', figsize=(10,5), title='Top 10 Countries on Netflix')
```

**Which countries contribute the most content?**

The United States contributes the largest number of titles which is at 2818, followed by India which is 1/3 compared to the United States, the United Kingdom, and other countries. This reflects both Hollywood’s global influence and Netflix’s major investment in Indian content.

```python
df['type'].value_counts()
```

```python
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6,6), title='Movies vs. TV Shows')
```

**What’s the balance between movies and TV shows?**

Movies make up the majority of Netflix’s catalog at almost 70%, but TV shows represent a substantial share. This balance highlights Netflix’s dual strategy of offering both feature films and long-form series.

```python
df['director'].value_counts().head(10)
```

```python
df['director'].value_counts().head(10).plot(kind='bar', figsize=(10,5), title='Top 10 Directors on Netflix')
```

**Who are the most frequent directors on Netflix?**

The chart reveals that a few directors, such as Raúl Campos and Jan Suter, appear frequently in the dataset, likely because of their work on multiple stand-up comedy specials. Most other directors have only one or two titles on the platform.
